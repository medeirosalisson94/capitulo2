```python
#Importar arquivo shapefile
pts_amos = gpd.read_file(r'C:\\Users\\Alisson Medeiros\\tese_cap2\\RF_brejos\\shp\\pres_aus.shp')

# Converter o GeoDataFrame para um ee.FeatureCollection
def gdf_to_ee_fc(gdf):
    features = []
    for _, row in gdf.iterrows():
        geom = row.geometry
        if geom.is_empty:
            continue
        coords = list(geom.coords)
        ee_geom = ee.Geometry.Point(coords[0])  # Pega o primeiro ponto (caso haja multipontos)
        feature = ee.Feature(ee_geom, row.drop('geometry').to_dict())
        features.append(feature)
    return ee.FeatureCollection(features)

pontos_ee = gdf_to_ee_fc(pts_amos)
```

```python
#RECORTAR DADOS IPSL-CM6A-LR_ssp126_2021-2040

# Caminhos para a pasta de entrada (shape)
input_folder = 'C:\\Users\\Alisson Medeiros\\tese_cap2\\shps\\recorte_dados_Intersect_BR.shp'
# Caminhos para a pasta de saída (rasters recortados)
output_folder = 'C:\\Users\\Alisson Medeiros\\tese_cap4\\RF_brejos\\bioc_IPSL-CM6A_rec'
# Carrega o arquivo shape
shapefile = os.path.join(input_folder)
gdf = gpd.read_file(shapefile)
# Lista os arquivos raster na pasta de entrada
raster_folder = 'C:\\Users\\Alisson Medeiros\\tese_cap4\\RF_brejos\\bioc_IPSL-CM6A'
raster_files = [f for f in os.listdir(raster_folder) if f.endswith('.tif')]
# Itera sobre os arquivos raster de precipitação
for raster_file in raster_files:
    raster_path = os.path.join(raster_folder, raster_file)
    # Abre o arquivo raster
    with rst.open(raster_path) as src:
        # Reprojeta o shapefile para a projeção do raster
        gdf_reprojected = gdf.to_crs(src.crs)
        # Extrai as áreas de interesse do raster com base no shapefile
        out_image, out_transform = mask(src, gdf_reprojected.geometry, crop=True)
        # Copia os metadados do raster original
        out_meta = src.meta.copy()
        # Atualiza os metadados com as informações da extração
        out_meta.update({"driver": "GTiff",
                         "height": out_image.shape[1],
                         "width": out_image.shape[2],
                         "transform": out_transform})
        # Define o nome do arquivo de saída
        output_file = os.path.splitext(raster_file)[0] + "wc2.1_30s_bioc_IPSL-CM6A-LR_ssp126_2021-2040.tif"
        output_path = os.path.join(output_folder, output_file)
        # Grava o raster recortado no arquivo de saída
        with rst.open(output_path, "w", **out_meta) as dest:
            dest.write(out_image)
            
        print(f"Arquivo '{raster_file}' recortado e salvo como '{output_file}'.")
```

**ALGUMAS VARIÁVEIS NÃO SEGUEM DISTRIBUIÇÃO NORMA**

Por este motivo usaremos o Teste de Kruskal-Wallis

```python
#Teste de Kruskal-Wallis
from scipy.stats import kruskal

# Kruskal-Wallis para precipitação
h_stat_prec, p_value_prec = kruskal(temp_prec2['prec_ssp126_2021-2040'], temp_prec2['prec_ssp585_2021-2040'])
print(f"Kruskal-Wallis para precipitação: H = {h_stat_prec}, p = {p_value_prec}")

# Kruskal-Wallis para temperatura
h_stat_temp, p_value_temp = kruskal(temp_prec2['temp_med_ssp126_2021-2040'], temp_prec2['temp_med_ssp585_2021-2040'])
print(f"Kruskal-Wallis para temperatura: H = {h_stat_temp}, p = {p_value_temp}")
```

**Kruskal-Wallis: Se o valor-p for menor ou igual ao nível de significância (0,05), a hipótese nula pode ser rejeitada.**

Hipótese nula: as medianas dos grupos são iguais

Valor de p para precipitação: p = 0.84 > 0.05

Valor de p para temperatura: p = 0.4666432562941186 > 0.05

*CONCLUSÃO: Para precipitação e temperatura, os resultados indicam que **não há diferença significativa** entre os modelos SSP126 e SSP585 no período de 2021-2040.*

Fator de Inflação de Variância (VIF)

O Fator de Inflação de Variância (VIF) é uma métrica estatística usada para avaliar multicolinearidade e orientar a seleção de variáveis. Ele indica o grau de relacionamento linear de cada variável independente com as outras variáveis ​​independentes, e altos valores de VIF podem ser evidências de multicolinearidade.

Normalmente, quando os valores de VIF excedem 5 (Valores de VIF maiores que 5 indicam multicolinearidade moderada) ou 10 (Valores de VIF maiores que 10 indicam multicolinearidade grave), isso sugere que a variável tem uma forte correlação com outras variáveis, potencialmente comprometendo a estabilidade e a interpretabilidade do modelo. Neste tutorial, um critério de valores de VIF menores que 10 foi usado para seleção de variáveis. As 6 variáveis ​​a seguir foram selecionadas com base no VIF.

```python
# Filtrar variáveis preditoras com base no  Fator de Inflação de Variância (VIF)
def filtrar_var_por_vif(pvals_df, threshold=10):

    original_columns = pvals_df.columns.tolist()
    remaining_columns = original_columns[:]
    
    while True:
        vif_data = pvals_df[remaining_columns]
        vif_values = [
            variance_inflation_factor(vif_data.values, i)
            for i in range(vif_data.shape[1])
        ]

        max_vif_index = vif_values.index(max(vif_values))
        max_vif = max(vif_values)

        if max_vif < threshold:
            break

        print(f"Removing '{remaining_columns[max_vif_index]}' with VIF {max_vif:.2f}")

        del remaining_columns[max_vif_index]
        
    filtered_data = pvals_df[remaining_columns]
    bands = filtered_data.columns.tolist()
    print("Bands:", bands)

    return filtered_data, bands
```

```python
filtered_pvals_df, bands = filtrar_var_por_vif(pvals_df)
```

```python
#Importar bibliotecas
import matplotlib.pyplot as plt
import seaborn as sns

#Correlação de Spearman das variáveis selecionadas
corrS_s = filtered_pvals_df.corr(method = 'spearman')

#Plotar gráfico
plt.figure(figsize=(12, 10))
sns.heatmap(corrS_s, annot = True, fmt=".2f", linewidths=0.5,); #fmt=".2f" reduz o número de casas, linewidths=0.5 ajusta espaçamento entre as células,
plt.title("Matriz de Correlação de Spearman das variáveis preditoras selecionadas via VIF - ssp585")
plt.tight_layout()
plt.savefig('Spearman_heatmap_plot_variáveis_selecionadas_VIF_- ssp585.png')
plt.show()
```
