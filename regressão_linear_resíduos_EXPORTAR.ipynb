```python
#Calcular dados para obter a regressão linear
#Plotar gráfico de regressão linear
import pandas as pd
import seaborn as sns

#Importar dados das amostras
pvals_GEE2 = pd.read_csv("C:\\Users\\Alisson Medeiros\\tese_cap4\\RF_brejos\\ee-chart.csv", sep = ',')

#Eliminar colunas desnecessárias
pvals_GEE2 = pvals_GEE2.rename(columns={'class_pred' : 'Valores_normalizados_preditos_pelo_modelo_RF',
                                   'med_var_no' : 'Valores_normalizados_dos_pontos_amostrais'})
pvals_GEE2

#nome variável = smf.ols('variável dependente ~ variável independente', data=datarame).fit()
regre_lin = smf.ols('Valores_normalizados_dos_pontos_amostrais ~ Valores_normalizados_preditos_pelo_modelo_RF', data=pvals_GEE2).fit()
print(regre_lin.summary())
```

```python
#Criar DataFrame com dados de coef da tabela acima
coef = pd.DataFrame(regre_lin.params)
coef.columns=['Coeficientes']
coef
```

```python
#Plotar gráfico de regressão linear
import numpy as np
plt.scatter(y=pvals_GEE2.Valores_normalizados_dos_pontos_amostrais, x=pvals_GEE2.Valores_normalizados_preditos_pelo_modelo_RF, color='blue', s=50, alpha=0.6)
X_plot = np.linspace(min(pvals_GEE2.Valores_normalizados_preditos_pelo_modelo_RF), max(pvals_GEE2.Valores_normalizados_dos_pontos_amostrais), len(pvals_GEE2.Valores_normalizados_dos_pontos_amostrais))
#X_plot = np.linspace(0,1000)
plt.plot(X_plot, X_plot*regre_lin.params[1] + regre_lin.params[0], color='r')
plt.title('Regressão linear')
plt.ylabel('Valores normalizados dos pontos amostrais')
plt.xlabel('Valores normalizados preditos pelo modelo RF')
#plt.savefig('Regressão lienar.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```

TESTE DE NORMALIDADE DOS RESÍDUOS

Fazer os testes de normalidade dos resíduos validam ou não a regressão linear. Se os resíduos não tiverem uma distribuição normal, a regressão não é válida.

Se p > 0.05, distribuição normal.
Se p <=0.05, a distribuição não é normal

```python
residuos = regre_lin.resid
residuos
```

```python
#Teste de Shapiro-Wilk
import scipy
from scipy import stats
stats.shapiro(residuos)

#pvalue=0.02262873575091362 < que 0.05. Distribuição NÃO É normal
```

```python
#Teste de Lilliefors

#Importar bibliotecas
!pip install statsmodels -q
import statsmodels
from statsmodels.stats.diagnostic import lilliefors

#Teste de Lilliefors (Kolmogorov_Sminorv) para mudanças florestais
statsmodels.stats.diagnostic.lilliefors(residuos, dist = 'norm')

#No teste Kolmogorov_Sminorv 0.4274572367165397, p > 0.05, ou seja, distribuição normal
```

```python
#Teste Anderson-Darling
stats.anderson(residuos, 'norm')

#No teste Anderson-Darling, o valor de p (0.74) > 0.05, ou seja, é uma distribuição normal
```

RMSE

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

# 2. Definir variáveis
y_true = pvals_GEE['Valores normalizados dos pontos amostrais']  # Valores reais (variável dependente)
y_pred = pvals_GEE['Valores normalizados preditos pelo modelo RF']  # Valores previstos (variável independente)

# 3. Calcular o RMSE
mse = mean_squared_error(y_true, y_pred)  # Calcula o erro quadrático médio (MSE)
rmse = np.sqrt(mse)  # Tira a raiz quadrada para obter o RMSE

print(f"RMSE: {rmse}")
```
